```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
def func(ga, gb, b):
    g1 = (1-b) * (ga + np.log(1-b))
    g2 = b * (gb + np.log(b))
    return g1 + g2
```

```python
x = np.linspace(0.001, 0.999, 1000)
g = func(0.6, 1, x)

pos_min = np.argmin(g)
g_min, x_min = g[pos_min], x[pos_min]

fig, ax = plt.subplots()
ax.plot(x, g)
ax.axvline(x_min, 0, g_min, linestyle='--')
ax.axhline(g_min, 0, x_min, linestyle='--')
ax.axhline(1, 0, 1, linestyle='--')
ax.set_ylim(bottom=0)
ax.set_xticks([0, x_min, 1])
ax.set_xticklabels(['0', r'$b_{eq}$', '1'])
ax.set_yticks([g_min, 0.6, 1])
ax.set_yticklabels([r'$G_{min}$', r'$G_a$', r'$G_b$'])
ax.set_xlim(0, 1)
ax.set_ylabel('Gibbs free energy')
ax.set_xlabel('Composition')

fig.set_size_inches(3,3)
fig.tight_layout()
fig.savefig('gibbs_profile.png', dpi=600, transparent=True)
```
