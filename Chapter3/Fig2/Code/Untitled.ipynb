```python
import sys
sys.path.append('../../../MembraneQuant')
from membranequant import *
%matplotlib notebook
```

### Normal polarity

```python
x = np.arange(0, 100)
A = (error_func(x, 52, 10)[::-1] + 1) / 2
P = (error_func(x, 52, 10) + 1) / 2

fig, ax = plt.subplots()
ax.plot(x, A, c='tab:red')
ax.plot(x, P, c='dodgerblue')
ax.set_xticks([])
ax.set_yticks([])
# ax.set_ylim([0, 1.1])
ax.set_ylabel('Concentration')
ax.set_xlabel('Position')
fig.set_size_inches(2, 1.7)
fig.tight_layout()
fig.savefig('normal_polarity.png', dpi=600, transparent=True)
```
