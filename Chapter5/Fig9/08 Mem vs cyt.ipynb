```python
import sys

sys.path.append('../../MembraneQuant')
from membranequant import *
import seaborn as sns

%matplotlib notebook
```

```python
df = pd.read_csv('res1_df.csv')
# df = pd.read_csv('res3_df.csv')
```

### Total membrane concentration

```python
df_filtered = df[df.Line.isin(['n2', 'nwg201', 'od58', 'nwg369'])]
fig, ax = plt.subplots()
sns.scatterplot(data=df_filtered, x='Cyt', y='Mem_tot', hue='Genotype', ax=ax)
ax.set_xlabel('Cytoplasmic concentration (a.u.)')
ax.set_ylabel('Membrane concentration (a.u.)')
ax.legend().set_title('')
plt.setp(ax.get_legend().get_texts(), fontsize='7')
ax.tick_params(axis='both', labelsize=8)
fig.set_size_inches(3, 3)
fig.tight_layout()
fig.savefig('Figs/cyt vs mem total.png', dpi=600, transparent=True)
```

### Posterior membrane concentration

```python
fig, ax = plt.subplots()
sns.scatterplot(data=df, x='Cyt', y='Mem_post', hue='Genotype', ax=ax)
ax.set_xlabel('Cytoplasmic concentration')
ax.set_ylabel('Posterior membrane concentration')
ax.set_xlim(right=7000)
ax.legend().set_title('')
plt.setp(ax.get_legend().get_texts(), fontsize=7)
ax.tick_params(axis='both', labelsize=8)
fig.set_size_inches(4,4)
fig.tight_layout()
```

```python
fig, ax = plt.subplots()
x = df['Cyt'] / 1000
y = df['Mem_tot'] / 1000
ax.scatter(x[df.Line == 'nwg201'], y[df.Line == 'nwg201'], facecolors='tab:blue', edgecolors='k')
ax.scatter(x[df.Line == 'nwg369'], y[df.Line == 'nwg369'], facecolors='tab:orange', edgecolors='k')
ax.scatter(x[df.Line == 'n2'], y[df.Line == 'n2'], facecolors='w', edgecolors='k')
ax.set_xlabel('Cytoplasmic concentration (a.u.)')
ax.set_ylabel('Membrane concentration (a.u.)')
ax.tick_params(axis='both', labelsize=8)

ax.fill_between([0, 4.2], [-0.5, -0.5], [12.5, 12.5], alpha=0.1, color='tab:blue', zorder=0)
ax.fill_between([0, 6.2], [-0.5, -0.5], [11.5, 11.5], alpha=0.1, color='tab:orange', zorder=0)

fig.set_size_inches(3.5, 3.5)
fig.tight_layout()
```
