```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv('Brood size assay for analysis.csv')
```

```python
df
```

```python
order = ['N2', 'Wild type', 'L50R', 'C56S', 'L109R', 'C56S, L109R', 'L50R, L109R', 'GCN4', 'C56S, GCN4', 
         'L109R, GCN4', 'GCN4w', 'L109R, GCN4w']
```

```python
fig, ax = plt.subplots()

sns.stripplot(data=df, x='Genotype', y='Lethality 5/11/21', ax=ax, order=order, clip_on=False)
ax.set_ylim(0, 1)
ax.set_ylabel('Embryonic lethality')
ax.set_xlabel('')
ax.set_zorder(-100)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_horizontalalignment('right')

fig.set_size_inches(8, 4)
fig.tight_layout()
fig.savefig('lethality.png', dpi=600, transparent=True)
```

```python
fig, ax = plt.subplots()

sns.stripplot(data=df, x='Genotype', y='Sterility 6/11/21', ax=ax, order=order, clip_on=False)
ax.set_ylim(0, 1)
ax.set_ylabel('Adult sterility')
ax.set_xlabel('')
ax.set_zorder(-100)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_horizontalalignment('right')

fig.set_size_inches(8, 4)
fig.tight_layout()
fig.savefig('sterility.png', dpi=600, transparent=True)
```
