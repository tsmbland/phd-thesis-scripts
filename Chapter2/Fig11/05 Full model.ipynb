```python
import sys
sys.path.append('../../MembraneQuant')
from membranequant import *
%matplotlib notebook
```

### Cytoplasmic reference

```python
x = np.arange(0, 50)
cyt_ref = error_func(x, 25, 5) + 1 + np.arange(0, 50) / 70
cyt_ref /= max(cyt_ref)

fig, ax = plt.subplots()
ax.plot(x,cyt_ref)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim([-0.1, 1.1])
ax.set_xlabel('Position (y)')
ax.set_ylabel('Norm. intensity')
fig.set_size_inches(1.5,1.5)
fig.tight_layout()
fig.savefig('Figs_full_model/cyt_ref.png', dpi=600, transparent=True)
```

### Cytoplasmic concentrations

```python
x = np.arange(0, 400)
cyts = 0.4 * np.ones(400)

fig, ax = plt.subplots()
ax.plot(x,cyts)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim([0, 1.1])
ax.set_xlabel('Position (x)')
ax.set_ylabel('Amplitude')
fig.set_size_inches(1.5,1.5)
fig.tight_layout()
fig.savefig('Figs_full_model/cyts.png', dpi=600, transparent=True)
```

### Cytoplasmic component

```python
cyt_component = np.expand_dims(cyt_ref, -1) * np.expand_dims(cyts, 0)

fig, ax = plt.subplots()
ax.imshow(cyt_component, cmap='viridis', vmin=0, vmax=1.25)
# ax.axis('off')
ax.set_xticks([])
ax.set_yticks([])
fig.set_size_inches(5,1)
fig.tight_layout()
fig.savefig('Figs_full_model/cyt_component.png', dpi=600, transparent=True)
```

### Membrane reference

```python
x = np.arange(0, 50)
mem_ref = gaus(x, 25, 2) + 0.1 * error_func(x, 25, 5) + 0.1 + np.arange(0, 50) / 500
mem_ref /= max(mem_ref)

fig, ax = plt.subplots()
ax.plot(x,mem_ref)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim([-0.1, 1.1])
ax.set_xlabel('Position (y)')
ax.set_ylabel('Norm. intensity')
fig.set_size_inches(1.5,1.5)
fig.tight_layout()
fig.savefig('Figs_full_model/mem_ref.png', dpi=600, transparent=True)
```

### Membrane concentrations

```python
x = np.arange(0, 400)
mems = (error_func(x, 320, 20) + error_func(x, 320, 20)[::-1] + 2) / 2

fig, ax = plt.subplots()
ax.plot(x,mems)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim([0, 1.1])
ax.set_xlabel('Position (x)')
ax.set_ylabel('Amplitude')
fig.set_size_inches(1.5,1.5)
fig.tight_layout()
fig.savefig('Figs_full_model/mems.png', dpi=600, transparent=True)
```

### Membrane component

```python
mem_component = np.expand_dims(mem_ref, -1) * np.expand_dims(mems, 0)

fig, ax = plt.subplots()
ax.imshow(mem_component, cmap='viridis', vmin=0, vmax=1.25)
ax.set_xticks([])
ax.set_yticks([])
fig.set_size_inches(5,1)
fig.tight_layout()
fig.savefig('Figs_full_model/mem_component.png', dpi=600, transparent=True)
```

### Full model

```python
fig, ax = plt.subplots()
ax.imshow(cyt_component + mem_component, cmap='viridis', vmin=0, vmax=1.25)
ax.set_xticks([])
ax.set_yticks([])
fig.set_size_inches(5,1)
fig.tight_layout()
fig.savefig('Figs_full_model/full_model.png', dpi=600, transparent=True)
```
