```python
def func(path, name, vmin=-200, vmax=2000):
    img = load_image(path + '/af_corrected.tif')
    roi = np.loadtxt(path + '/ROI.txt')
    img_rotated = rotated_embryo(img, roi, 250, 200, 3)
    
    fig, ax = plt.subplots()
    ax.imshow(img_rotated, cmap='gray', vmin=vmin, vmax=vmax)
    ax.axis('off')
    fig.set_size_inches(3,3)
    
    save_img_jpeg(img_rotated, name + '.jpeg', cmin=vmin, cmax=vmax, cmap='gray')
```
