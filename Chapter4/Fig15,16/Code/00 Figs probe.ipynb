### Focal planes

```python
path = '/Users/blandt/Desktop/Data/Tomm20/Probe/190916_nwg231/06 zstack'
img = load_image(glob.glob(path + '/*561 SP 630-75*')[0])
roi = np.loadtxt(path + '/ROI.txt')
rotated = [rotated_embryo(i, roi, 350, 270, 3) for i in img]
max_projection = np.max(np.array(rotated), axis=0)
```

```python
positions = [10, 30, 50]

fig, ax = plt.subplots(1, 4)
ax[0].imshow(rotated[positions[0]], cmap='gray')
ax[1].imshow(rotated[positions[1]], cmap='gray')
ax[2].imshow(rotated[positions[2]], cmap='gray')
ax[3].imshow(max_projection, cmap='gray')
[a.axis('off') for a in ax]
fig.set_size_inches(8, 2)
fig.tight_layout()
fig.subplots_adjust(wspace=0.03)
fig.savefig('Figs/focal_planes.png', dpi=600, transparent=True)
```
