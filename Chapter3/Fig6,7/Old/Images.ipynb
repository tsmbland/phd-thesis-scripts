```python
from scipy.ndimage.filters import gaussian_filter
import ipywidgets as widgets

fig, ax = plt.subplots()
@widgets.interact(i=(0, len(images_wt) - 1, 1))
def update(i=0):
    ax.clear()
    ax.imshow(images_wt[i], cmap='gray')
    ax.axis('off')
    ax.set_title('%s, %s' % (i, dosages_wt[i]))
```

```python
fig, ax = plt.subplots()
@widgets.interact(i=(0, len(images_c56s) - 1, 1))
def update(i=0):
    ax.clear()
    ax.imshow(images_c56s[i], cmap='gray')
    ax.axis('off')
    ax.set_title(i)
```

```python
from scipy.ndimage import median_filter

def func(ax, img):
    print(img.shape)
    cyt = np.mean(img[75:125, 125:175])
    ax.imshow(median_filter(img, size=3), cmap='gray', vmin=0, vmax=3 * cyt)
    ax.axis('off')

    
    
fig, axes = plt.subplots(2, 3)

func(axes[0][0], images_wt[8])
func(axes[0][1], images_wt[17])
func(axes[0][2], images_wt[29])

func(axes[1][0], images_c56s[10])
func(axes[1][1], images_c56s[19])
func(axes[1][2], images_c56s[25])

fig.set_size_inches(7, 3)
```
