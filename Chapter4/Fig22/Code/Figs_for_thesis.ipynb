```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv('Brood size assay for analysis.csv')
df['Lethality 5/11/21'] = df['Lethality 5/11/21'] * 100
df['Sterility 6/11/21'] = df['Sterility 6/11/21'] * 100
```

```python
order = ['N2', 'Wild type', 'L50R', 'L109R', 'C56S', 'L50R, L109R', 'C56S, L109R']
labels = ['N2 (untagged)', 'Wild type', 'L50R', 'L109R', 'C56S', 'L50R + L109R', 'C56S + L109R']
palette=['0.7', 'tab:blue', 'tab:orange', 'tab:green', 'tab:red', '0.7', '0.7']
```

```python
fig, ax = plt.subplots()
palette=['0.7', 'tab:blue', 'tab:orange', 'tab:green', 'tab:red', '0.7', '0.7']
sns.stripplot(data=df, x='Genotype', y ='Lethality 5/11/21', order=order, linewidth=1, palette=palette, clip_on=False)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_horizontalalignment('right')
    
ax.set_xticklabels(labels)
ax.set_ylim(0, 100)
ax.set_ylabel('% Embryonic lethality')
ax.set_xlabel('')
ax.set_zorder(-100)
ax.tick_params(axis='both', labelsize=8)
fig.set_size_inches(3, 3)
fig.tight_layout()
fig.savefig('Figs_for_thesis/lethality.png', dpi=600, transparent=True)
```

```python
fig, ax = plt.subplots()
sns.stripplot(data=df, x='Genotype', y ='Sterility 6/11/21', order=order, linewidth=1, palette=palette, clip_on=False)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_horizontalalignment('right')
    
ax.set_xticklabels(labels)
ax.set_ylim(0, 100)
ax.set_ylabel('% Adult sterility')
ax.set_xlabel('')
ax.set_zorder(-100)
ax.tick_params(axis='both', labelsize=8)
fig.set_size_inches(3, 3)
fig.tight_layout()
fig.savefig('Figs_for_thesis/sterility.png', dpi=600, transparent=True)
```
