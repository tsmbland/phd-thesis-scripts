```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import copy
import matplotlib.ticker as ticker
import matplotlib as mpl
import itertools
import pandas as pd
%matplotlib notebook

datadirec = '/Volumes/lab-goehringn/home/users/blandt/ModelData/dimer_kinetic_pPAR_dimerisation_only_e_beta_scheme4_v2/'
```

```python
def paramspace_fig(n, iteration):
    path = datadirec + str(n)
    
    fig, ax = plt.subplots()
#     df = pd.read_csv(path + '/%s.csv' % iteration, header=None)
#     ax.scatter(df[0], df[1], c=df[2], s=10) 

    for i in range(iteration + 1):
        df = pd.read_csv(path + '/%s.csv' % i, header=None)
        ax.scatter(df[0], df[1], c=df[2], cmap='tab10', s=10, vmax=10) 
        
    ax.set_xlim(0.01, 0.99)
    ax.set_xticks([0.01, 0.25, 0.5, 0.75, 0.99])
    ax.set_ylim(0, 1)
    ax.tick_params(axis='both', labelsize=8)
    ax.set_xlabel(r'$\epsilon$')
    ax.set_ylabel(r'$\beta$')

    fig.set_size_inches(2.5, 2.5)
    fig.tight_layout()
    return fig, ax
```

```python
fig, ax = paramspace_fig(29, 0)
ax.set_title('Iteration 1', fontsize=12)
fig.subplots_adjust(top=0.88)
fig.savefig('Figs_method/iteration1.png', dpi=600, transparent=True)
```
