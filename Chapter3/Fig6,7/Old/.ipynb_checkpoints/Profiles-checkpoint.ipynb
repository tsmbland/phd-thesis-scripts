```python
import sys
sys.path.append('../../MembraneQuant')
from membranequant import *
%matplotlib notebook
```

```python
path = '/Users/blandt/Desktop/Analysis/Experiments/e2001__par2_rundown/'
direcs = direcslist(path, 1)

straights = []
for d in direcs:
    img = load_image(d + '/af_corrected.tif')
    roi = np.loadtxt(d + '/ROI_fitted.txt')
    straights.append(straighten(img, roi, 50))
```

```python
profs = np.zeros([len(direcs), 50])
profs_norm = np.zeros([len(direcs), 50])
for i, s in enumerate(straights):
    profs[i] = bounded_mean_2d(s, [0, 1])
    profs_norm[i] = profs[i] / np.mean(profs[i][45:])
```

```python
df = pd.read_csv('df.csv')
print(df)
```

```python
ax = plt.axes(projection='3d')
condition = (df['Dosage'] > 500) * (df['Line'] == 'nwg246')
profs_norm_thresh = profs_norm[condition]
dosages_thresh = df['Dosage'][condition]
for d, p in zip(dosages_thresh, profs_norm_thresh):
    ax.plot(np.arange(50), np.ones(50) * d/ 1000, p, alpha=0.8)
    
ax.set_ylabel('\n\nTotal PAR-2 (a.u.)')
ax.set_zlabel('\nNormalised intensity')
ax.set_xlabel('\nPosition (pixels)')
```
