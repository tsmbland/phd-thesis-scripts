```python
import sys
sys.path.append('../../MembraneQuant')
from membranequant import *
%matplotlib notebook
import matplotlib.patches as patches
```

```python
def merge(img_c, img_m):
    g = (img_c - np.min(img_c)) / (np.max(img_c) - np.min(img_c))
    r = (img_m - np.min(img_m)) / (np.max(img_m) - np.min(img_m))
    b = np.sqrt((r ** 2) + (g ** 2)) 
#     b = (b - np.min(b)) / (np.max(b) - np.min(b))
    merge = np.dstack((r, g, b))
    return merge
```

```python
rect1 = patches.Rectangle((ybox[0], xbox[0]), ybox[1] - ybox[0], xbox[1] - xbox[0], 
                         linewidth=1, edgecolor='tab:orange', facecolor='none')
rect2 = patches.Rectangle((ybox[0], xbox[0]), ybox[1] - ybox[0], xbox[1] - xbox[0], 
                         linewidth=1, edgecolor='tab:orange', facecolor='none')
rect3 = patches.Rectangle((ybox[0], xbox[0]), ybox[1] - ybox[0], xbox[1] - xbox[0], 
                         linewidth=1, edgecolor='tab:orange', facecolor='none')

fig, ax = plt.subplots(1, 3)
ax[0].imshow(rotatedR, cmap='gray')
ax[1].imshow(rotatedG, cmap='gray')
ax[2].imshow(merge(rotatedG, rotatedR))
ax[0].add_patch(rect1)
ax[1].add_patch(rect2)
ax[2].add_patch(rect3)
ax[0].set_title('Tomm20::GBP::mKate', fontsize=10, c='m')
ax[1].set_title('GFP::PAR-2', fontsize=10, c='c')
ax[2].set_title('Merge', fontsize=10)
[a.axis('off') for a in ax]
fig.set_size_inches(6, 2)
fig.tight_layout()
fig.savefig('Figs/coloc_par2.png', dpi=600, transparent=True)
```
